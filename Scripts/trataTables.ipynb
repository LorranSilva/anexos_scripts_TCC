```python
# Pivot Crimes

df = pd.read_csv('../Data/crimes.csv',  sep=';', encoding='latin1')

df['ano'] = pd.to_numeric(df['ano'], errors='coerce')
indice_coluna_referencia = df.columns.get_loc('regiao')
colunas_crimes = df.columns[indice_coluna_referencia + 1:]

data = []

for ano in range(2019, 2023):
    for cityId in CITIESCODE:
        count = 1
        for crime in colunas_crimes:
            if crime not in ['fase', 'registro_ocorrencias', 'total_furtos', 'total_roubos']:
                filtro = (df['ano'] == ano) & (df['mcirc'] == cityId)
                ocorrencias = df[filtro][crime].sum()
                data.append({'nk_crime': count,'ano': ano, 'cidadeId': cityId, 'crime': crime, 'ocorrencias': int(ocorrencias)})
                count += 1

df = pd.DataFrame(data)
df.to_csv("../Data/crimesTratado.csv", index=False)
```
